```python
## Rename the column.

import pandas as pd

df.rename(columns={'availability 365': 'days_booked'}, inplace=True)

print("DataFrame after renaming the column:")
df.head()


```

```python
## Convert all column names to lowercase and replace the spaces with an underscore "_"

df.columns = df.columns.str.lower().str.replace(' ', '_')

print("DataFrame after converting column names:")
df.head()

```

```python
## Remove the dollar sign and comma from the columns. If necessary, convert these two columns to the appropriate data type.

df['price'] = df['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)
df['service_fee'] = df['service_fee'].replace({'\$': '', ',': ''}, regex=True).astype(float)

print("DataFrame after removing the dollar sign and comma:")
df.head()
```

### Task 4: Exploratory Data Analysis (Any Tool)

- List the count of various room types avaliable in the dataset.
- Which room type has the most strict cancellation policy?
- List the average price per neighborhood group, and highlight the most expensive neighborhood to rent from.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
## List the count of various room types avaliable with Airbnb

room_type_counts = df['room_type'].value_counts()

print("Count of various room types available with Airbnb:")
room_type_counts
```

```python
## Which room type adheres to more strict cancellation policy


cancellation_policy_counts = df.groupby('room_type')['cancellation_policy'].value_counts()


most_strict_policy = cancellation_policy_counts.groupby(level=0).idxmax()


print("Room type with the most strict cancellation policy:")
most_strict_policy

```

```python
cancellation_policy_counts = df.groupby(['room_type', 'cancellation_policy']).size().reset_index(name='count') 
                                        
pivot_table = pd.pivot_table(cancellation_policy_counts, values='count', index='room_type', columns='cancellation_policy', fill_value=0)
pivot_table_percentage = pivot_table.div(pivot_table.sum(axis=1), axis=0) * 100

room_type_with_strict_policy = pivot_table_percentage.idxmax(axis=0)['strict']

print("Room type with more strict cancellation policy:")
room_type_with_strict_policy

```

```python
## List the prices by neighborhood group and also mention which is the most expensive neighborhood group for rentals

prices_by_neighbourhood = df.groupby('neighbourhood_group')['price'].mean()

print("Prices by neighbourhood group:")
prices_by_neighbourhood

```

```python
most_expensive_neighbourhood_group = prices_by_neighbourhood.idxmax()

print("\nMost expensive neighbourhood group for rentals:")
most_expensive_neighbourhood_group
```

```python
# Create the box and whisker chart that showcases the price distribution of all listings split by room type.

plt.figure(figsize=(10, 6))
sns.boxplot(x='price', y='room_type', data=df, palette='viridis')
plt.xlabel('Price')
plt.ylabel('Room Type')
plt.title('Price Distribution by Room Type')
plt.show()
```
