## Task 2b: Data Cleaning (Python)

- Check for missing values in the dataframe and display the count in ascending order. **If the values are missing, impute the values as per the datatype of the columns.**
- Check whether there are any duplicate values in the dataframe and, if present, remove them.
- Display the total number of records in the dataframe before and after removing the duplicates.

```python
## Check for missing values in the dataframe and display the count in ascending order.

import pandas as pd

missing_values_count = df.isnull().sum().sort_values(ascending=True)

missing_values_count

```

```python
for col in df.columns:
    if df[col].dtype == 'O':
        print(col)
        df[col].fillna(value=df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(value=df[col].median(), inplace=True)

df.isnull().sum()
```

```python
## Check whether there are any duplicate values in the dataframe and if present remove them.

df.duplicated().sum()
```

```python
## Check whether there are any duplicate values in the dataframe and if present remove them.

duplicates = df.duplicated()

duplicates

```

```python
df.drop_duplicates(inplace=True)


print("\nDataFrame after removing duplicates:")
df

```

## Task 3: Data Transformation (Any Tool)


- Rename the column `availability 365` to `days_booked`
- Convert all column names to lowercase and replace the spaces in the column names with an underscore "_".
- Remove the dollar sign and comma from the columns `price` and `service_fee`. If necessary, convert these two columns to the appropriate data type.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
## Rename the column.

import pandas as pd

df.rename(columns={'availability 365': 'days_booked'}, inplace=True)

print("DataFrame after renaming the column:")
df.head()
```

```python
## Convert all column names to lowercase and replace the spaces with an underscore "_"

df.columns = df.columns.str.lower().str.replace(' ', '_')

print("DataFrame after converting column names:")
df.head()
```

```python
## Remove the dollar sign and comma from the columns. If necessary, convert these two columns to the appropriate data type.

def remove_dollar_sign(value):
    if pd.isna(value):
        return np.NaN
    else:
        return float(value.replace( "$","").replace(",","").replace(" ",'"'))
                                          
```

```python
df['price'] = df['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)
df['service_fee'] = df['service_fee'].replace({'\$': '', ',': ''}, regex=True).astype(float)

print("DataFrame after removing the dollar sign and comma:")
df.head()
```

```python
## Which room type adheres to more strict cancellation policy

df_group_two = df[df['cancellation_policy'] == 'strict']

room_type_counts = df_group_two['room_type'].value_counts()

print(room_type_counts)

```

```python
cancellation_policy_counts = df.groupby(['room_type', 'cancellation_policy']).size().reset_index(name='count') 
                                        
pivot_table = pd.pivot_table(cancellation_policy_counts, values='count', index='room_type', columns='cancellation_policy', fill_value=0)
pivot_table_percentage = pivot_table.div(pivot_table.sum(axis=1), axis=0) * 100

room_type_with_strict_policy = pivot_table_percentage.idxmax(axis=0)['strict']

print("Room type with more strict cancellation policy:")
room_type_with_strict_policy
```

## Task 5a: Data Visualization (Any Tool)

- Create a horizontal bar chart to display the top 10 most expensive neighborhoods in the dataset.
    - Create another chart with the 10 cheapest neighborhoods in the dataset.
- Create a box and whisker chart that showcases the price distribution of all listings split by room type.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
# Create the horizontal bar chart to display the top 10 most expensive neighborhoods in the dataset.
import matplotlib.pyplot as plt

grp2 = df.groupby('neighbourhood')['price'].sum().sort_values(ascending=False)
top_neighborhoods = grp2.head(10)

ax = plt.gca()
ax.get_xaxis().get_major_formatter().set_scientific(False)

top_neighborhoods.plot(kind='barh', color='green') 
plt.xlabel("Total Price")
plt.ylabel("Neighborhood")
plt.title("Top 10 Neighborhoods by Total Price")
plt.tight_layout() 
plt.show()
```

```python
# Create another chart with the 10 cheapest neighborhoods in the dataset.

import seaborn as sns
import matplotlib.pyplot as plt

df_group_three = df[df['minimum_nights'] <= 10]

neighbourhood_counts = df_group_three['neighbourhood_group'].value_counts()

sns.barplot(x=neighbourhood_counts.values,
            y=neighbourhood_counts.index,
            orient='h',
            palette='viridis')

plt.xlabel("Count")
plt.ylabel("Neighbourhood Group")
plt.title("Neighbourhood Group Distribution for Minimum Nights <= 10")
plt.tight_layout()
plt.show()
```

```python
# Create the box and whisker chart that showcases the price distribution of all listings split by room type.


import seaborn as sns

dfl = df.groupby(['room_type', 'price']).agg(mean_price=("price", 'mean'))
dfl = dfl.reset_index()

# plot barplot
sns.barplot(x="room_type", y="mean_price", data=dfl)

```

```python
#Create a line chart to showcase the total amount of listings available per year.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df["construction_year_counts"] = df["construction_year"].value_counts()

sns.lineplot(x="construction_year", y="construction_year_counts", data=df)

plt.xlabel("Construction Year")
plt.ylabel("Listing Counts")
plt.title("Total Amount of Listings Available per Year")
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

df_filtered = df[['instant_bookable', 'construction_year', 'calculated_host_listings_count', 'days_booked']]

# Using .assign() to create the new 'year' column without modifying the original DataFrame
df_filtered = df_filtered.assign(year=pd.to_datetime(df_filtered['construction_year'], format='%Y').dt.year)

yearly_listings = df_filtered.groupby('year')['calculated_host_listings_count'].sum()

# Creating the line chart
plt.figure(figsize=(10, 6))

plt.plot(yearly_listings.index, yearly_listings.values, marker='o', linestyle='-', color='b')

plt.xlabel('Year')
plt.ylabel('Total Listings')
plt.title('Total Listings Available per Year')
plt.grid(True)
plt.xticks(yearly_listings.index, rotation=45)
plt.tight_layout()
plt.show()
```
